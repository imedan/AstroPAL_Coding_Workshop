```python
import numpy as np
import time
from to_precision import to_precision # /!\ You'll have to install to_precision to use this code!
import decimal

ddir='' #data directory
tdir='' #table data directory, specifically for LaTeX_Tableizer

def val(num,roundto):
    if roundto==0:
        return num
    else:
        try:
            return to_precision(num, roundto, 'std')
        except ValueError:
            return num

def err(num):
    try:
        e=to_precision(num, 2, 'std')
        if e[-1]=='0':
            return to_precision(e, 2, 'std')
        else:
            return e
    except ValueError:
        return num

#roundtable auto-recognizes errors and rounds the vals to correct digits!
def roundtable(dat,labels,unknownerr='11'): #array with rows and columns,list, formatted like readlines(). Can set unknownerr (unknown error) to 2-digit number with second 
    types=['e' if 'err' in d else 'v' for d in labels] # place value the place to which you'll round. Default: Round to 1's place.
    table=[]
    for d in dat:
        dd=[]
        for i in range(len(types)):
            if 'nan' in d[i]:
                v=r'\nodata'
            elif types[i]=='v' and i==len(types)-1: #last value in data table, therefore value has no error
                v=d[i]
            elif types[i]=='e' and 'nan' not in d[i]: #known error
                v=err(d[i])
            elif types[i]=='v' and types[i+1]=='e' and 'nan' not in d[i] and 'nan' not in d[i+1]: #value with known error
                v=valround(d[i],err(d[i+1]))
            elif types[i]=='v' and types[i+1]=='e' and 'nan' not in d[i] and 'nan' in d[i+1]: #value with unknown error
                v=valround(d[i],unknownerr)
            else:
                v=d[i]
            dd.append(v)
        #print(dd)
        table.append(dd)
    print(table[0])
    print('...\n')
    return(table)

def valround(value,er): #input the output from err()
    digits=abs(decimal.Decimal(er).as_tuple().exponent)
    #print digits
    if 'nan' in str(digits):
            v=r'\nodata'
    elif digits==0:
        erraboveground=len(er.strip('.'))-2
        if int(float(value))==0:
            if erraboveground==0:
                v='0.'
            else:
                v='0'
        else:
            whole=len(str(int(float(value))).strip('-'))
            sigfigs=whole-erraboveground
            v=str(to_precision(float(value), sigfigs, 'std'))
    else:
        if float(value)==0:
            whole=1
        elif int(float(value))==0:
            whole=0
        else:
            whole=len(str(int(float(value))).strip('-'))
        sigfigs=whole+digits
        vpre=str(to_precision(float(value), sigfigs, 'std'))
        if len(vpre[vpre.index('.'):])>len(er[er.index('.'):]): #limit trailing zeros
            v=vpre[:-(len(vpre[vpre.index('.'):])-len(er[er.index('.'):]))]
        else:
            v=vpre
    return v

def cutoff(dat,column,dec,mode='normal'): #dat from roundtable,index of column you want to round, decimal cutoff for that column
    col=[d[column] for d in dat] #column to fix
    newcol=[]
    for c in col:
        if mode=='normal':
            a=c.strip()
        if mode=='radec':
            RA=c.strip().split()
            a=RA[2]
        try:
            ind=a.index('.')
            if len(a[ind+1:])>dec: #counting numbers past the decimal point
                if dec==0:
                    newa=('%.0f') %(float(a)+float(a)/abs(float(a))*1.0*10.**(-1.*dec-2.))
                    if newa[-1]==0:
                        newa+='.'
                elif dec==1:
                    newa=('%.1f') %(float(a)+float(a)/abs(float(a))*1.0*10.**(-1.*dec-2.))
                elif dec==2:
                    newa=('%.2f') %(float(a)+float(a)/abs(float(a))*1.0*10.**(-1.*dec-2.))
                elif dec==3:
                    newa=('%.3f') %(float(a)+float(a)/abs(float(a))*1.0*10.**(-1.*dec-2.))
                elif dec==4:
                    newa=('%.4f') %(float(a)+float(a)/abs(float(a))*1.0*10.**(-1.*dec-2.))
                else:
                    print("Add more or check your error limits. :P")
            else:
                newa=a
        except ValueError:
            newa=a
        if mode=='normal':
            newcol.append(newa)
        if mode=='radec':
            newra=RA[0]+' '+RA[1]+' '+newa
            newcol.append(newra)
    #print newcol
    newdat=[dat[i][:column]+[newcol[i],]+dat[i][column+1:] for i in range(len(dat))]
    return newdat

def racutoff(ra,dec): #ra or dec as 00 00 00, decimal...bad nomenclature here, it's not Dec dec, it's decimal dec
    RA=ra.strip().split()
    a=RA[2]
    try:
        ind=a.index('.')
        if len(a[ind+1:])>dec: #counting numbers past the .
            #print 'try',a
            if dec==0:
                newa=('%.0f') %(float(a)+float(a)/abs(float(a))*1.0*10.**(-1.*dec-2.))
                if newa[-1]==0:
                    newa+='.'
            elif dec==1:
                newa=('%.1f') %(float(a)+float(a)/abs(float(a))*1.0*10.**(-1.*dec-2.))
            elif dec==2:
                newa=('%.2f') %(float(a)+float(a)/abs(float(a))*1.0*10.**(-1.*dec-2.))
            elif dec==3:
                newa=('%.3f') %(float(a)+float(a)/abs(float(a))*1.0*10.**(-1.*dec-2.))
            elif dec==4:
                newa=('%.4f') %(float(a)+float(a)/abs(float(a))*1.0*10.**(-1.*dec-2.))
            else:
                print("Add more or check your error limits. :P")
        else:
            newa=a
    except ValueError:
        newa=a
    newra=RA[0]+' '+RA[1]+' '+newa
    return newra

def opendat(dir,filename): #dir,'filename'. For opening a data file. Can then send through roundtable.
    f=open(dir+filename,'r')
    dat=f.readlines()
    f.close()
    labels=dat[0][0:-1].split('\t')
    dat2=[[a.strip('\n') for a in d.split('\t')] for d in dat][1:]
    return [dat2,labels]
```

## Textify

This code takes data files written in the format from writedat() with a commented-out header line at the top and all data tab-separated. If you place value columns next to error columns and label them in the format "val" and "valerr", the code automatically rounds everything such that uncertainties have 2 significant figures and values have as many decimal places as the rounded uncertainties. Decimal-place limits can be (clunkily :s) imposed by the cutoff() function, which operates on a whole data list.

The main functions to use are roundtable() and cutoff(). Function opendat() can be used to load data (e.g. perform the dat = readlines() step).

```python
# Simple case: You have a data file you wrote in the correct format, and you want to
# turn it directly into a LaTeX deluxe table.

dat,labels=opendat(ddir,'SampleData_Kinematics.dat')
print(labels)

f=open(tdir+'kinematics_table.dat','w')
print(' \t '.join(labels),file=f)
for d in roundtable(dat,labels):
    print(' \t '.join(d),file=f)
f.close()

print('Sent on '+time.strftime("%Y-%m-%d %H:%M")+'.') #Helps you track how updated your tables are
```

```python
# Slightly more complex case: I want to cut off the decimal places in my age column.

dat,labels=opendat(ddir,'SampleData_ClusterProperties.dat')
print(labels)

dat2=roundtable(dat,labels) #prepare MOST values through roundtable
dat3=cutoff(dat2,2,0) #perform decimal cutoffs on column 2. Leave 0 decimal places.
dat4=cutoff(dat3,3,0) #perform decimal cutoffs on column 3. Leave 0 decimal places.
dat5=[[d[i].replace('0.','0') if i==2 or i==3 else d[i] for i in range(len(d))] for d in dat4] #clean up .0 and 0. stuff
print(dat5[0]) #check how it looks

f=open(tdir+'ClusterProperties_table.dat','w')
print('\t'.join(labels),file=f)
for d in dat5:
    print('\t'.join(d),file=f)
f.close()

print('Sent on '+time.strftime("%Y-%m-%d %H:%M")+'.')
```
