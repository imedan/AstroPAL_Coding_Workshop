```python
#for AAStex deluxetable

import numpy as np

tdir='' #table data directory

def table(dir,filename): #dir,'filename'. Auto name correct, err treatment, and even/uneven errs!
    f=open(dir+filename,'r')
    dat=f.readlines()
    f.close()
    
    leader=dat[0].split()
    pminds=[i-1 for i in range(len(leader)) if 'err' in leader[i]]
    cells=len(leader)-len(pminds)
    formatt='l'+r'c'*(cells-1)
    del leader
    
    print(r'\begin{deluxetable}{'+formatt+r'}'+'[h] %table [h] for here, table* [t] for full-page table')
    print(r'\tablecaption{Title \label{'+filename+r'}}')
    print(r'\tablecolumns{'+str(cells)+'}')
    #print(r'\tablenum{2}')
    print(r'\tablewidth{0pt}')
    
    dattext=[]
    for j in range(len(dat)):
        d=dat[j]
        dd=sum([[a.strip(),' & '] for a in d.split('\t')],[]) #line = [data,'&',data,'&'...]

        ind2=[pminds[i] for i in range(len(pminds)-1) if pminds[i+1]-pminds[i]==1] #handle double errors
        ind22=sum([[a,a+1] for a in ind2],[]) #for ref for ind1
        ind1=[a for a in pminds if a not in ind22]

        for i in ind1:
            dd[i*2+1]=r' $\pm$ ' #replaces the '&' between data and error
        for i in ind2:
            if dd[(i+1)*2].strip()!=dd[(i+2)*2].strip(): #if double errors don't match
                dd[i*2+1]=r'$^{+'
                dd[(i+1)*2+1]=r'}_{-'
                dd[(i+2)*2+1]=r'}$ & '
            elif dd[(i+1)*2].strip()==dd[(i+2)*2].strip(): #if double errors do match
                dd[i*2+1]=r' $\pm$ '
                dd[(i+1)*2+1]=''
                dd[(i+2)*2]=''
        dd[-1]=r' \\'
        whole=''.join(dd)
        
        fixwhole=whole.replace('#','').replace('41A','41 A').replace('41B','41 B').replace('cg30','CG 30').replace('_irs4',' IRS 4').replace('_x','').replace('_a04',' (a04)').replace('_s1','').replace('_s2','').replace('_narrow',' (nar)').replace('_n2',' (2)').replace(r'$\pm$ \nodata','').replace(r'$^{+\nodata}_{-\nodata}$','').replace('KWW','KWW ').replace('XRS','XRS ').replace('blank','').replace('IRAS','IRAS ').replace('VdBH9b','VdBH 9b ') #.replace('pha',r'PH$\alpha$ ')
        if j==0:
            #column names
            heads=[a.strip() for a in fixwhole.strip(' \\').split('&')]
            header=r'\tablehead{\colhead{'+r'} & \colhead{'.join(heads)+r'}}'
            print(header)
            print(r'\startdata')
        else:
            #data
            print(fixwhole)
            dattext.append(fixwhole)
        #if j==0:
        #    print r'\hline'
    
    #print(r'\tablenotetext{a}{Footnote here.}' #paired with footnoter r'\tablenotemark{a}' )
    #print(r'\tablecomments{General table note here.}')
    print(r'\enddata')
    print(r'\end{deluxetable}')
    
    return dattext
```

## Make and update pretty LaTeX tables with ease!

This function table() takes the nice rounded data you made in Textify and gives you an almost-ready-to-use LaTeX deluxetable! No more hassle! No more repeated manual nitpicking! :) All you have to do is copy-paste and prettify the column headers. Then every time you update your data, you run this code again and just copy the data section between startdata/ and enddata/. :)

You can add manipulations using the fact that all the data is put into the variable $t$.

```python
# Simple case: Your table is ready to go, and you just want to tableize it!
# Copy-paste the table into your LaTeX file, prettify the headers, and go.

t=table(tdir,'ClusterProperties_table.dat')
```
